### 4.	Utilizar los siguientes métodos para cada columna que contiene missing values

```python
titanicMDcpy = pandas.DataFrame()
titanicMDcpy["PassengerId"] = titanicMD["PassengerId"]
titanicMDcpy["Survived"] = titanicMD["Survived"]
titanicMDcpy["Pclass"] = titanicMD["Pclass"]
titanicMDcpy["Name"] = titanicMD["Name"]
titanicMDcpy["Ticket"] = titanicMD["Ticket"]
titanicMDcpy["Cabin"] = titanicMD["Cabin"]
```

```python
# Age
# I.mean
avAge = math.floor(titanicMD["Age"].mean())
titanicMDcpy["Age"] = titanicMD["Age"].fillna(avAge).astype(int)
```

```python
# SibSp
# I.mode
modeSibSp = int(titanicMD["SibSp"].mode())
titanicMDcpy["SibSp"] = titanicMD["SibSp"].fillna(modeSibSp)
```

```python
# Parch
# I.mode
modeParch = int(titanicMD["Parch"].mode())
titanicMDcpy["Parch"] = titanicMD["Parch"].fillna(modeParch).astype(int)
```

```python
# Fare
# I.mean
avFare = int(titanicMD["Fare"].mean())
titanicMDcpy["Fare"] = titanicMD["Fare"].fillna(avFare).astype(int)
```

```python
# Embarked
# I.mode
nll =  titanicMD["Embarked"].isnull()

titanicMD['EmS'] = LabelEncoder.fit_transform(titanicMD['Embarked'])
titanicMD["EmS"] = titanicMD["EmS"].where(nll != True, numpy.NaN)
modeEmbarked = int(titanicMD["EmS"].mode())
titanicMD["Em"] = titanicMD["EmS"].fillna(modeEmbarked).astype(int)
titanicMDcpy["Embarked"] = LabelEncoder.inverse_transform(titanicMD["Em"])
```

### 1.	Luego del pre-procesamiento de la data con Missing Values, normalice las columnas numéricas por los métodos: 

* Standarization

* MinMaxScaling

* MaxAbsScaler


```python
# Standarization
ageS = StandardScaler.fit_transform(numpy.array(titanicMDcpy["Age"]).reshape(-1,1))
fareS = StandardScaler.fit_transform(numpy.array(titanicMDcpy["Fare"]).reshape(-1,1))
AgeS = StandardScaler.fit_transform(numpy.array(titanic["Age"]).reshape(-1,1))
FareS = StandardScaler.fit_transform(numpy.array(titanic["Fare"]).reshape(-1,1))

print("\nStandarization\n")
print("Fare:")
print(fareS)
print("\nFare2:")
print(FareS)
print("\nAge:")
print(ageS)
print("\nAge2:")
print(AgeS)
```

```python
# MinMaxScaling
ageMnMx = MinMaxScaler.fit_transform(numpy.array(titanicMDcpy["Age"]).reshape(-1,1))
fareMnMx = MinMaxScaler.fit_transform(numpy.array(titanicMDcpy["Fare"]).reshape(-1,1))
AgeMnMx = MinMaxScaler.fit_transform(numpy.array(titanic["Age"]).reshape(-1,1))
FareMnMx = MinMaxScaler.fit_transform(numpy.array(titanic["Fare"]).reshape(-1,1))

print("\nMinMaxScaling\n")
print("Fare:")
print(fareMnMx)
print("\nFare2:")
print(FareMnMx)
print("\nAge:")
print(ageMnMx)
print("\nAge2:")
print(AgeMnMx)
```

```python
# MaxAbsScaler
ageMa = MaxAbsScaler.fit_transform(numpy.array(titanicMDcpy["Age"]).reshape(-1,1))
fareMa = MaxAbsScaler.fit_transform(numpy.array(titanicMDcpy["Fare"]).reshape(-1,1))
AgeMa = MaxAbsScaler.fit_transform(numpy.array(titanic["Age"]).reshape(-1,1))
FareMa = MaxAbsScaler.fit_transform(numpy.array(titanic["Fare"]).reshape(-1,1))

print("\nMaxAbsScaler\n")
print("Fare:")
print(fareMa)
print("\nFare2:")
print(FareMa)
print("\nAge:")
print(ageMa)
print("\nAge2:")
print(AgeMa)
```
