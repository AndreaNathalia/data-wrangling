### 3. Cuántos superhéroes miden más de 2 metros (200 cms)

```python
p3 = "SELECT COUNT(Id) FROM heroes WHERE Height > 200"
pysqldf(p3)
```

### 4. Cuántos superhéroes son humanos y miden más de 2 metros (200 cms)?

```python
p4 = "SELECT * FROM heroes WHERE Race IN ('Human') AND Height > 200"
pysqldf(p4)
```

R// 12

### 5. Cuántos superhéroes pesan más de 100 lbs y menos de 200 lbs?

```python
p5 = "SELECT Name FROM heroes WHERE Weight BETWEEN 100 AND 200"
#p5 = "SELECT * FROM heroes WHERE Race IN(1,3,5,7)"
pysqldf(p5)
```

### 6. Cuántos superhéroes tienen los ojos de color azul o rojo?

```python
p6 = "SELECT Name, `Eye color` FROM heroes WHERE `Eye color` IN ('red', 'blue')"
pysqldf(p6)
```

R// 271

### 7. Cuántos superhéroes son Human, Mutant y tienen el pelo color Verde o son Vampiros con pelo Negro

```python
p7 = "SELECT Name, Race, `Hair color` FROM heroes  WHERE Race IN ('Human','Mutant') AND `Hair color` IN ('Green') OR Race IN ('Vampire') AND `Hair color` IN ('Black')"
pysqldf(p7)
```

R// 5

### 8. Cuál es el nombre del primer superhéroe si ordeno la tabla por raza en orden descendente?

### 9. Cuántos superhéroes son de género masculino y cuántas de género femenino?
